```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

prices = pd.read_csv('prices.txt', sep=r'\s+')
print(prices.shape)
prices.head()
```

```python
plt.plot(prices.iloc[:, 0])   # ALGO column
plt.title("ALGO price")
plt.xlabel("Day")
plt.ylabel("Price")

plt.plot(prices.iloc[:, 1])   # AAVE column
plt.title("AAVE price")
plt.xlabel("Day")
plt.ylabel("Price")

plt.show()
```

```python
# figure with 1 row and 2 columns of plots
fig, axs = plt.subplots(1, 2)

axs[0].plot(prices.iloc[:, 0])   # ALGO column
axs[0].set_title("ALGO price")
axs[0].set_xlabel("Day")
axs[0].set_ylabel("Price")

axs[1].plot(prices.iloc[:, 1])   # AAVE column
axs[1].set_title("AAVE price")
axs[1].set_xlabel("Day")
axs[1].set_ylabel("Price")

plt.tight_layout()
plt.show()
```

```python
sns.clustermap(prices.corr('spearman'), cmap='Blues', figsize=(20, 20));
```

```python
corr = prices.corr(method='spearman')

# Only look at the upper triangle (avoid duplicates)
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

# Find highly correlated pairs
high_corr = (
    upper.stack()
         .sort_values(key=abs, ascending=False)
)

# Display correlations with absolute value > 0.8
print(high_corr[abs(high_corr) > 0.8])
```

```python
fig, axes = plt.subplots(6, 2, figsize=(14, 18))
axes = axes.flatten()

for ax, ((stock1, stock2), corr_value) in zip(axes, high_corr.items()):
    ax.plot(prices[stock1] / prices[stock1].iloc[0], label=stock1)
    ax.plot(prices[stock2] / prices[stock2].iloc[0], label=stock2)

    ax.set_title(f"{stock1} vs {stock2} ρ = {corr_value:.3f}")
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```
