```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

prices = pd.read_csv('prices.txt', sep=r'\s+')
print(prices.shape)
prices.head()
```

```python
fig, axes = plt.subplots(26, 2, figsize=(14, 50))
axes = axes.flatten()

for i, stock in enumerate(prices.columns):
    axes[i].plot(prices[stock] / prices[stock].iloc[0])
    axes[i].set_title(stock)

fig.delaxes(axes[-1])
plt.tight_layout()
plt.show()
```

```python
sns.clustermap(prices.corr('spearman'), cmap='Blues', figsize=(20, 20));
```

```python
corr = prices.corr(method='spearman')

# Only look at the upper triangle (avoid duplicates)
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

# Find highly correlated pairs
high_corr = (
    upper.stack()
         .sort_values(key=abs, ascending=False)
)

# Display correlations with absolute value > 0.8
print(high_corr[abs(high_corr) > 0.8])
```

```python
fig, axes = plt.subplots(6, 2, figsize=(14, 18))
axes = axes.flatten()

for ax, ((stock1, stock2), corr_value) in zip(axes, high_corr.items()):
    # spread
    p1 = prices[stock1] / prices[stock1].iloc[0]
    p2 = prices[stock2] / prices[stock2].iloc[0]

    spread = p1 - p2

    plt.figure(figsize=(10,4))
    plt.plot(spread)
    plt.axhline(spread.mean(), color="red", linestyle="--")
    plt.title("Spread")
    plt.show()

    ax.set_title(f"{stock1} vs {stock2} ρ = {corr_value:.3f}")
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(6, 2, figsize=(14, 18))
axes = axes.flatten()

for axes, stock in zip(axes, prices.columns):
    p1 = prices[stock1] / prices[stock1].iloc[0]
    p2 = prices[stock2] / prices[stock2].iloc[0]

    spread = p1 - p2

    axes.hist(spread, bins=40, density=True, alpha=0.7, edgecolor='black')
    axes.set_title(f"{stock1} - {stock2}")

plt.tight_layout()
plt.show()
```
